### **Head recursion**

```python
count = 0
def func(count):
    if count == 4:
        return
    print("Varun")
    count += 1
func(count)

# Time Complexity: O(n) - The function makes a linear number of recursive calls until it reaches the base case, where n is the value of count at which the function stops calling itself.
# Space Complexity: O(n) - Each recursive call adds a new frame to the call stack, and in the worst case, there can be n frames in the stack until the base case is reached.

# Head Recursion is a type of recursion where the recursive call is made at the beginning of the function, before any other operations or computations. In head recursion, the function calls itself before performing any other tasks, and the base case is typically checked after the recursive call. This can lead to a large number of recursive calls and may result in a stack overflow if the depth of recursion exceeds the maximum limit.
```

### **Tail recursion**

```python
count = 0
def func(count):
    if count == 4:
        return
    count += 1
func(count)
print("Varun")

# Time Complexity: O(n) - Similar to head recursion, the function makes a linear number of recursive calls until it reaches the base case, where n is the value of count at which the function stops calling itself.
# Space Complexity: O(n) - Each recursive call adds a new frame to the call stack, and in the worst case, there can be n frames in the stack until the base case is reached.

# Tail Recursion is a special case of recursion where the recursive call is the last operation in the function. In tail recursion, there are no further computations or operations after the recursive call. This allows for optimization by the compiler or interpreter, as it can reuse the same stack frame for each recursive call, preventing stack overflow and improving performance.
```

### **Recursion Using Parameter**

```python
# [Print x, n times]
def func(x, n):
    if n == 0:
        return
    print(x)
    func(x, n - 1)

func(5,4)
```

```python
# [Print 1 to n] head recursion
def func(i, n):
    if i > n:
        return
    print(i)
    func(i + 1, n)

func(1, 5)
```

```python
# [Print 1 to n] tail recursion
def func(i, n):
    if i > n:
        return
    func(i + 1, n)
    print(i)

func(1, 5)
```

```python
# [Print n to 1] head recursion
def func(n):
    if n == 0:
        return
    print(n)
    func(n - 1)
func(5)
```
