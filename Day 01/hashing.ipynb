## Question 1


```python
n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

# constraints
# 1) 1 <= n[i] <= 10
# 2) n can have 10^8 elements
# 3) m can have 10^8 elements
```

### **Approach 1: Brute Force**

```python
for num in m:
    count = 0
    for i in n:
        if i == num:
            count += 1
    print(f"{num}:{count}", end=" ")

# Time Complexity = O(n*m) where n is the length of list n and m is the length of list m
# Space Complexity = O(1) as we are using only a constant amount of space to store the count variable.
```

###  **Approch 2: Optimal**

```python
hash_list = [0] * 11
for num in n:
    hash_list[num] += 1
for num in m:
    if num < 1 or num > 10:
        print(0)
    else:
        print(hash_list[num])

# Time Complexity = O(n+m) where n is the length of list n and m is the length of list m
# Space Complexity = O(1) as we are using only a constant amount of space to store the hash_list variable.
```

### **Approch 3: Dictionary**

```python
dict_hash = {}
for num in n:
    if num in dict_hash:
        dict_hash[num] += 1
    else:
        dict_hash[num] = 1
for num in m:
    if num in dict_hash:
        print(dict_hash[num])
    else:
        print(0)
        
```

## Question 2

```python
s = "azyxyyzaaaa"
q = ["d", "a", "y", "x"]

# constraints
# 1) "a" <= s[i] <= "z"
```

```python
hash_list = [0] * 26
for char in s:
    ascii_value = ord(char)
    index = ascii_value - 97
    hash_list[index] += 1
for char in q:
    ascii_value = ord(char)
    index = ascii_value - 97
    print(hash_list[index])

# Time Complexity = O(n+m) where n is the length of string s and m is the length of list q
# Space Complexity = O(1) as we are using only a constant amount of space to store the hash_list variable.
```
