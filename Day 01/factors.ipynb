#### Step 1

```python
num = 20
result = []
for i in range(1, num + 1):
    if num % i == 0:
        result.append(i)
print(result)
# Time complexity: O(n)
# Space complexity: O(n)
```

#### Step 2

```python
num = 20
result = []
for i in range(1, num // 2):
    if num % i == 0:
        result.append(i)
result.append(num)
print(result)

# Time complexity: O(n)
# Space complexity: O(n)
```

#### Step 3

```python
# Optimal solution
num = 36
result = []
from math import sqrt
for i in range(1, int(sqrt(num)) + 1):
    if num % i == 0:
        result.append(i)
        if i != num // i:
            result.append(num // i)
result.sort()
print(result)

# Time complexity: O(sqrt(n)) + O(n log n)
# Space complexity: O(k) where k is the number of factors
```
